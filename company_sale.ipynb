# Visualizing Company Sale data

### Step 1. Import the necessary libraries

```python
import pandas as pd
from pylab import *
from collections import Counter
import numpy as np
# set this so the graphs open internally
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://github.com/thieu1995/csv-files/blob/main/data/matplotlib/company_sales_data.csv).

```python
data=pd.read_csv('https://raw.githubusercontent.com/thieu1995/csv-files/main/data/matplotlib/company_sales_data.csv')
```

```python
data
```

### Exercise 1: Read Total profit of all months and show it using a line plot

![img](img/01.png)

```python
figure(figsize=(7,5))
plot(data['total_profit'])
yticks([100000,200000,300000,400000,500000])
xticks(list(np.arange(1,13,1)))
ylabel('Total profit')
xlabel('Month number')
title('Company profit per month')
show()
```

### Exercise 2: Get total profit of all months and show line plot with the following Style properties

![img](img/02.png)

```python
figure(figsize=(7,5))
plot(data['total_profit'],label = 'Profit data of last year', 
      color='r', marker='o', markerfacecolor='k', 
      linestyle='--', linewidth=3)
yticks([100000,200000,300000,400000,500000])
xticks(list(np.arange(1,13,1)))
ylabel('Total profit')
xlabel('Month number')
title('Company sale data of last year')
legend()
show()
```

### Exercise 3: Read all product sales data and show it  using a multiline plot

![img](img/03.png)

```python
sale_unit=data.loc[:,'facecream':'moisturizer']
plot(sale_unit, marker='o', linewidth=3)
yticks([1000,2000,4000,6000,8000,10000,12000,15000,18000])
xticks(list(np.arange(1,13,1)))
legend(['Face cream Sales Data','Face Wash Sales Data','ToothPaste Sales Data','bathingsoap Sales Data','shampoo Sales Data','moisturizer Sales Data'])
xlabel('Month Number')
ylabel('Number of units Sold')
title(' Sales data')
show()
```

### Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot
![img](img/04.png)

```python
sale_unit=data.loc[:,'toothpaste':'toothpaste']
scatter(np.arange(1,13),sale_unit)
yticks(np.arange(4500,8001,500))
xticks(list(np.arange(1,13,1)))
grid(True,linewidth= 1, linestyle="--")
legend(['ToothPaste Sales Data'])
xlabel('Month Number')
ylabel('Number of units Sold')
title(' Sales data')
show()
```

### Exercise 5: Read face cream and facewash product sales data and show it using the bar chart

![img](img/05.png)

```python
faceCremSalesData   = data['facecream'].tolist()
faceWashSalesData   = data['facewash'].tolist()
bar(np.arange(0.75,12.75,1), faceCremSalesData, width= 0.25, label = 'Face Cream sales data', align='edge')
bar(np.arange(1.25,13.25,1), faceWashSalesData, width= -0.25, label = 'Face Wash sales data', align='edge')
xlabel('Month Number')
ylabel('Sales units in number')
legend(loc='upper left')
title(' Sales data')
xticks(list(np.arange(1,13,1)))
grid(True, linewidth= 1, linestyle="--")
title('Facewash and facecream sales data')
show()
```

### Exercise 6: Read sales data of bathing soap of all months and show it using a bar chart. Save this plot to your hard disk

![img](img/06.png)

```python
bathingsoapSalesData   = data['bathingsoap'].tolist()
bar(list(np.arange(1,13,1)), bathingsoapSalesData)
xlabel('Month Number')
ylabel('Sales units in number')
title(' Sales data')
xticks(list(np.arange(1,13,1)))
grid(True, linewidth= 1, linestyle="--")
title('bathingsoap sales data')
show()
```

### Exercise 7: Read the total profit of each month and show it using the histogram to see the most common profit ranges

![img](img/07.png)

```python
profitList = data['total_profit'].tolist()
labels = ['low', 'average', 'Good', 'Best']
profit_range = [150000, 175000, 200000, 225000, 250000, 300000, 350000]
hist(profitList, profit_range, label = 'Profit data')
xlabel('profit range in dollar')
ylabel('Actual Profit in dollar')
legend(loc='upper left')
xticks(profit_range)
title('Profit data')
show()
```

### Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart

![img](img/08.png)

```python
labels = ['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
salesData   = [data['facecream'].sum(), data['facewash'].sum(), data['toothpaste'].sum(), data['bathingsoap'].sum(), data['shampoo'].sum(), data['moisturizer'].sum()]
axis("equal")
pie(salesData, labels=labels, autopct='%1.1f%%')
legend(loc='lower right')
title('Sales data')
show()
```

### Exercise 9: Read Bathing soap facewash of all months and display it using the Subplot

![img](img/09.png)

```python
bathingsoap   = data['bathingsoap'].tolist()
faceWashSalesData   = data['facewash'].tolist()

f, axarr = plt.subplots(2, sharex=True)
axarr[0].plot(list(np.arange(1,13,1)), bathingsoap, label = 'Bathingsoap Sales Data', color='k', marker='o', linewidth=3)
axarr[0].set_title('Sales data of  a Bathingsoap')
axarr[1].plot(list(np.arange(1,13,1)), faceWashSalesData, label = 'Face Wash Sales Data', color='r', marker='o', linewidth=3)
axarr[1].set_title('Sales data of  a facewash')

xticks(list(np.arange(1,13,1)))
xlabel('Month Number')
ylabel('Sales units in number')
show()
```

### Exercise 10: Read all product sales data and show it using the stack plot

![img](img/10.png)

```python
faceCremSalesData   = data['facecream'].tolist()
faceWashSalesData   = data['facewash'].tolist()
toothPasteSalesData = data['toothpaste'].tolist()
bathingsoapSalesData   = data['bathingsoap'].tolist()
shampooSalesData   = data['shampoo'].tolist()
moisturizerSalesData = data['moisturizer'].tolist()

plot([],[],color='m', label='face Cream', linewidth=5)
plot([],[],color='c', label='Face wash', linewidth=5)
plot([],[],color='r', label='Tooth paste', linewidth=5)
plot([],[],color='k', label='Bathing soap', linewidth=5)
plot([],[],color='g', label='Shampoo', linewidth=5)
plot([],[],color='y', label='Moisturizer', linewidth=5)

stackplot(list(np.arange(1,13,1)), faceCremSalesData, faceWashSalesData, toothPasteSalesData, 
              bathingsoapSalesData, shampooSalesData, moisturizerSalesData, 
              colors=['m','c','r','k','g','y'])

xlabel('Month Number')
ylabel('Sales unints in Number')
title('Alll product sales data using stack plot')
legend(loc='upper left')
show()
```
